```julia
using LinearAlgebra
using Plots
```

**Excersize 3**

```julia
random_matrix = randn(Float16, 4, 4)

L, U = lu(random_matrix)

display(random_matrix)

plt0 = heatmap(L, c = cgrad([:black, :white]))
plt1 = heatmap(U, c = cgrad([:black, :white]))

display(plt0)
display(plt1)

display(L)
display(U)

determinant = 1

for i in 1:1:size(U, 1)
    determinant *= U[i, i] 
end

println("Determinant from L: ", determinant)
println("Determinant from Julia: ", det(random_matrix))
```

**Excersize 4**

```julia
random_matrix = randn(Float16, 4, 4)
L, U, p = lu(random_matrix)
p_matrix = random_matrix[p, :]

display(inv(random_matrix))
print("=")
display(inv(U) * inv(L) * p_matrix)
```

**Excersize 5**

```julia
random_matrix = randn(Float16, 4, 4)

L, U = lu(random_matrix)

display(transpose(random_matrix) * random_matrix) 
println(" = ") 
display(transpose(U)* transpose(L) * L * U)
```
