```python
import numpy as np
import os
from PIL import Image
import matplotlib.pyplot as plt
from keras.applications.vgg19 import VGG19
from keras.models import Sequential
from keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy, CategoricalCrossentropy
from tensorflow.keras.metrics import Accuracy
from keras.utils import to_categorical
import pandas as pd
from sklearn.metrics import classification_report
```

Load images

```python
grapes=[]

cat=0
for subdir in os.listdir("/content/drive/MyDrive/plants/grape"):
    counter=0
    for pic in os.listdir(os.path.join("/content/drive/MyDrive/plants/grape",subdir)):
        if(counter==200):
            break

        grapes.append([np.load(os.path.join("/content/drive/MyDrive/plants/grape",subdir,pic)),cat]) 
        counter+=1   
    
    cat += 1
```

```python
for x in range(len(grapes)):
    grapes[x][0]=Image.fromarray(grapes[x][0])
```

```python
counter=0
for x in grapes:
    x[0].save("/content/drive/MyDrive/temp/"+str(x[1])+"/"+str(counter)+".jpg")
    counter+=1
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=45,
    horizontal_flip=True,
    vertical_flip=True
    )
```

```python
import splitfolders as sf
sf.ratio("/content/drive/MyDrive/temp","/content/drive/MyDrive/plants/splitted",ratio=(.75,.15,.1))
```

fine tuning

```python
model = Sequential()
base=VGG19(include_top=False,input_shape=(256,256,3))
for layer in base.layers[:-1]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.add(Flatten())
model.add(Dense(2,"softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(),metrics=[Accuracy()],loss=CategoricalCrossentropy())
```

```python
gen=datagen.flow_from_directory("/content/drive/MyDrive/plants/splitted/train",class_mode='binary')
```

```python
x_train=[]
y_train=[]
for subdir in os.listdir("/content/drive/MyDrive/plants/splitted/train"):
    counter=0
    for pic in os.listdir(os.path.join("/content/drive/MyDrive/plants/splitted/train",subdir)):
        x_train.append([plt.imread(os.path.join("/content/drive/MyDrive/plants/splitted/train",subdir,pic))]) 
        y_train.append(subdir)
        counter+=1   
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
# x_val=np.array(x_val)
# y_val=np.array(y_val)
```

```python
len(x_train)
```

```python
x_train=np.squeeze(x_train)
# x_val=np.squeeze(x_val)
```

```python
# y_val=to_categorical(y_val)
y_train=to_categorical(y_train)
```

```python
y_train=y_train.reshape(-1,1)
# y_val=y_val.reshape(-1,1)
```

```python
model.fit(x=x_train,y=y_train,validation_split=0.2,epochs=10, shuffle=True)
```

```python
result=pd.DataFrame(model.history.history)
```

```python
result[["loss","val_loss"]].plot()
```

```python
result[["accuracy","val_accuracy"]].plot()
```

test

```python
x_test=[]
y_test=[]
for subdir in os.listdir("/content/drive/MyDrive/plants/splitted/test"):
    counter=0
    for pic in os.listdir(os.path.join("/content/drive/MyDrive/plants/splitted/test",subdir)):
        x_test.append([plt.imread(os.path.join("/content/drive/MyDrive/plants/splitted/test",subdir,pic))]) 
        y_test.append(int(subdir))
        counter+=1   
```

```python
x_test=np.squeeze(x_test)
```

```python
prediction=np.argmax(model.predict(x_test), axis=-1)
```
