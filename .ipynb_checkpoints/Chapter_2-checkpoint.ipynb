**Excersize 2-1**

```julia
using Plots

v = [1 2]
w = [4 6]
sum_vector = v + w

plot([0, v[1]], [0, v[2]], arrow=true, label="v")
plot!([v[1], v[1] + w[1]], [v[2], -w[2] + v[2]], arrow=true, label="w")
plot!([0, sum_vector[1]], [0, v[2] - w[2]], arrow=true, label="w + v", xlims=(-6, 6), ylims=(-6, 6), xlabel = "x", ylabel = "y",
                                title = "Vectors v, w and v + w")
```

```julia
sub = v - w

plot([0, v[1]], [0, v[2]], arrow=true, label="v")
plot!([0, w[1]], -[0, w[2]], arrow=true, label="w")
plot!([-sub[2], v[1]], [-w[2], v[2]], arrow=true, label="w - v", xlims=(-6, 6), ylims=(-6, 6), xlabel = "x", ylabel = "y",
                                title = "Vectors v, w and v - w")
```

**Excersize 2-2**

```julia
function norm(vector)
    return (sqrt(sum(vector .^2)))
end
```

```julia
x = [1 2 3]
norm(x)
```

**Excersize 2-3**

```julia
function unit_vector(vector)
    return (1 / sqrt(sum((vector .^ 2)))) * vector    # we can calculate the denumerator usgin norm(vector)
end
```

```julia
unit_vector([3, 2, 1])
```

**Excersize 2-4**

```julia
function vector_with_magnitude(magnitude, vector)
    return magnitude * unit_vector(vector)
end
```

```julia
vector_with_magnitude(2, [3, 2, 1])
```
