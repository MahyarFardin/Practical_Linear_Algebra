```julia
x = -1:.1:1
y = .√(1 .- x .^ 2)

# vertical concatination for adding negative data
x = vcat(x, x)
y = vcat(y, .-y)

# making a n*2 matrix
xy = hcat(x, y)

kernel = [1 .5 ; 0 .5]
y_transformed = xy * kernel

scatter(x, y)
scatter!(x, y_transformed[: , 2])
```

```julia
ϕ = -1:.1:1

x = -1:.1:1
y = sin.(x)
xy = hcat(x, y)
```

```julia
@userplot CirclePlot
@recipe function f(cp::CirclePlot)
    x, y, i = cp.args
    n = length(x)
    inds = circshift(1:n, 1 - i)
    linewidth --> range(0, 10, length = n)
    seriesalpha --> range(0, 1, length = n)
    aspect_ratio --> 1
    label --> false
    x[inds], y[inds]
end

anim = @animate for i in ϕ
    
    T = [(1 - i / 3) 0 ; 0 i]
    
    result = xy * T
    plot(result)
end
gif(anim, "animation.gif", fps = 15)
```
