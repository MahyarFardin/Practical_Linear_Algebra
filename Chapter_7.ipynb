```julia
x = -1:.1:1
y = .(1 .- x .^ 2)

# vertical concatination for adding negative data
x = vcat(x, x)
y = vcat(y, .-y)

# making a n*2 matrix
xy = hcat(x, y)

kernel = [1 .5 ; 0 .5]
y_transformed = xy * kernel

scatter(x, y)
scatter!(x, y_transformed[: , 2])
```
